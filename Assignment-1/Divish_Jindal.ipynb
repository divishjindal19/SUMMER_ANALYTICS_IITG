### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
median_mpg = data['mpg'].median()
data['hp_per_weight'] = data['horsepower'] / data['weight']
above_median_df = data[data['mpg'] > median_mpg]
max_ratio = above_median_df['hp_per_weight'].max()
best_ratio_cars = above_median_df[above_median_df['hp_per_weight'] == max_ratio]['name'].tolist()
print(f"Car(s) with the best horsepower-to-weight ratio among cars with above-median mpg: {best_ratio_cars}")
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
plt.figure(figsize=(10, 6))
avg_mpg_by_year_origin = data.groupby(['model_year', 'origin'])['mpg'].mean().unstack()
for origin in avg_mpg_by_year_origin.columns:
    plt.plot(avg_mpg_by_year_origin.index, avg_mpg_by_year_origin[origin], marker='o', label=origin)
plt.title('Evolution of Average MPG by Origin Over the Years')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
name_year_counts = data.groupby('name')['model_year'].nunique()
multi_year_names = name_year_counts[name_year_counts > 1]

mpg_std_by_name = data.groupby('name')['mpg'].std()
mpg_std_multi_year = mpg_std_by_name[multi_year_names.index]

consistent_cars = mpg_std_multi_year[mpg_std_multi_year < 1.0]
```

```python
if not consistent_cars.empty:
    lowest_std = consistent_cars.min()
    lowest_std_cars = consistent_cars[consistent_cars == lowest_std]

    mpg_mean_by_name = data.groupby('name')['mpg'].mean()

    result_df = pd.DataFrame({
        'name': lowest_std_cars.index,
        'std_mpg': lowest_std_cars.values,
        'num_appearances': multi_year_names[lowest_std_cars.index],
        'avg_mpg': mpg_mean_by_name[lowest_std_cars.index]
    })

    result_df = result_df.sort_values(by=['num_appearances', 'avg_mpg'], ascending=[False, False])

    print("Consistent car models (std_mpg < 1.0) with the lowest variation:")
    print(result_df[['name', 'num_appearances', 'avg_mpg']])
else:
    print("No car models meet the criteria (appear in multiple years with std_mpg < 1.0).")
```
